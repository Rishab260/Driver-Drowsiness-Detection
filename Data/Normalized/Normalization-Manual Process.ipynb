```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../Merged/Table_0.csv')
```

```python
df = df.iloc[:,2:]
```

```python
df_alert = df[df['State'] == 0]
```

```python
df_alert
```

```python
def divide_df(df):
    smaller_dfs = []
    n = 3
    for i in range(0, df.shape[0], n):
        smaller_dfs.append(df[i:i+n].copy())
    return smaller_dfs
```

```python
df_alert_list = divide_df(df_alert)
```

```python
df_alert_list
```

```python
for i,df in zip(range(len(df_alert_list)), df_alert_list):
    scaler = StandardScaler()
    print(df)
    df_alert_list[i] = pd.DataFrame(scaler.fit_transform(df.values), columns = df.columns, index = df.index)
    print(df_alert_list[i])
    
```

```python
df_alert_list
```

```python
df_merged_alert = pd.concat(df_alert_list,axis=0)
```

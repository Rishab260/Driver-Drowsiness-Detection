```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import glob
import os
```

```python
tables = glob.glob('..\\Merged_2\\*.csv')
```

```python
tables
```

```python
def split_3(df):
    smaller_dfs = []
    n = 3
    for i in range(0, df.shape[0], n):
        smaller_dfs.append(df[i:i+n].copy())
    return smaller_dfs
```

```python
for table in tables:
    dff = pd.read_csv(table)
    dff = dff.iloc[:,2:]
    df_list = split_3(dff)
    #print(df_list)
    for i,df in zip(range(len(df_list)), df_list):
        
        scaler = StandardScaler()
        df_temp = df.copy()
       

        #print(df)
        df_list[i] = pd.DataFrame(scaler.fit_transform(df.values), columns = df.columns, index = df.index)
        df_list[i].drop("State", axis=1, inplace = True)
        df_list[i].insert(4,'State',df_temp['State'].values)
        
   
        #df_list[i] = df
        print("\n")
        print(f"\n{i} DF_Temp\n\n",df_temp)
        print(f"\n{i} DF_Scaled\n\n",df_list[i])

    dff_n = pd.concat(df_list, axis=0, ignore_index = True)
    dff_n.to_csv(f"NormalizedData\\{os.path.basename(table)}")
```

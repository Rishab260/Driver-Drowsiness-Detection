```python
import os
import pandas as pd

# Set the path to the folder containing the CSV files
folder_path = '../Merged'

# Get a list of all the CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty list to store the normalized dataframes
normalized_dfs = []

# Loop over each CSV file and normalize the data
for csv_file in csv_files:
    # Read in the CSV file as a dataframe
    df = pd.read_csv(os.path.join(folder_path, csv_file))
    
    # Get the first three rows of the dataframe
    first_three_rows = df.head(3)
    
    # Calculate the mean and standard deviation of each feature
    means = first_three_rows.mean()
    stds = first_three_rows.std()
    
    # Use the means and standard deviations to normalize the rest of the data
    df.iloc[:, :-1] = (df.iloc[:, :-1] - means) / stds
    
    # Add the normalized dataframe to the list
    normalized_dfs.append(df)

# Concatenate the dataframes into a single feature matrix
X = pd.concat(normalized_dfs, ignore_index=True)
```

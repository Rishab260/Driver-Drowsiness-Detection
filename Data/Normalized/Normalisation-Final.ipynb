```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import glob
import os
```

```python
tables = glob.glob('..\\Merged\\*.csv')
```

```python
tables
```

```python
def split_3(df):
    smaller_dfs = []
    n = 3
    for i in range(0, df.shape[0], n):
        smaller_dfs.append(df[i:i+n].copy())
    return smaller_dfs
```

```python
df_list = []
for table in tables:
    dff = pd.read_csv(table)
    dff = dff.iloc[:,2:]
    dff = dff[(dff['State'] != 5) & (dff['State'] != 1) ]
    print(dff)
    
    scaler = StandardScaler()
    dff.iloc[:, :-1] = scaler.fit_transform(dff.iloc[:, :-1])
    df_list.append(dff)
Combined = pd.concat(df_list,axis=0, ignore_index=True)
```
